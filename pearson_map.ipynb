```python
import matplotlib.pyplot as plt
import seaborn as sns

# Rename columns for better visualization with subscripts using Matplotlib's math text
column_renames = {}
# Check if 'Eb (kV/cm)' exists and rename it
if 'Eb (kV/cm)' in train.columns:
    column_renames['Eb (kV/cm)'] = '$E_b$ (kV/cm)'
# Check if 'W' exists and rename it
if 'Wrec' in train.columns:
    column_renames['Wrec'] = '$W_{rec}$'

# Apply renaming if any changes are specified
if column_renames:
    train = train.rename(columns=column_renames)

# 1) Compute correlation
corr = train.corr()

# 2) Smaller figure size (boxes more compact)
plt.figure(figsize=(10, 8), dpi=600)  # reduced further from 16x14

sns.heatmap(
    corr,
    cmap="YlGnBu",
    annot=False,           # remove numbers
    linewidths=0.2,
    linecolor="white",
    square=True,
    cbar_kws={"shrink": 0.8}
)

# 3) Larger feature label font
plt.xticks(
    rotation=90,
    fontsize=14,           # larger font size
    fontweight='bold'
)
plt.yticks(
    rotation=0,
    fontsize=14,           # larger font size
    fontweight='bold'
)

plt.title("Pearson Correlation Heatmap", fontsize=18, pad=10)

plt.tight_layout()

# 4) Save as TIFF
plt.savefig(
    "pearson_heatmap.tiff",
    format="tiff",
    dpi=600,
    bbox_inches="tight"
)

plt.show()
```
